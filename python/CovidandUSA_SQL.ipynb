### Common Table Expressions

````python
%%sql
-- Creating the table holding valus for Milwaukee in Wisconsin
WITH 
wisconsin_2020 AS (select * from us_counties_2020 where state = 'Wisconsin' and county = 'Milwaukee'),
wisconsin_2021 AS (select * from us_counties_2021 where state = 'Wisconsin' and county = 'Milwaukee'),
wisconsin_2022 AS (select * from us_counties_2022 where state = 'Wisconsin' and county = 'Milwaukee'),
wisconsin_2023 AS (select * from us_counties_2023 where state = 'Wisconsin' and county = 'Milwaukee'),
wisconsin AS (select * from wisconsin_2020 UNION select * from wisconsin_2021 UNION select * from wisconsin_2022 UNION select * from wisconsin_2023)

-- Creating new columns to hold the dailycases and deaths number
select *,  
       COALESCE(cases - lag(cases) OVER(ORDER BY DATE), 0) as DailyCases,
       COALESCE(deaths - lag(deaths) OVER(ORDER BY DATE), 0) as DailDeaths
from wisconsin
order by date
````

### Answering the queries

### Top 10 Days with the highest number of deaths reported in IL

````python
%%sql
-- Creating the table holding valus for Illinois
WITH 
illinois_2020 AS (select * from us_counties_2020 where state = 'Illinois'),
illinois_2021 AS (select * from us_counties_2021 where state = 'Illinois'),
illinois_2022 AS (select * from us_counties_2022 where state = 'Illinois'),
illinois_2023 AS (select * from us_counties_2023 where state = 'Illinois'),
illinois AS (select * from illinois_2020 UNION select * from illinois_2021 UNION select * from illinois_2022 UNION select * from illinois_2023)

-- Creating new columns to hold the dailycases and deaths number
select *,  
       COALESCE(cases - lag(cases) OVER(ORDER BY COUNTY, DATE), 0) as DailyCases,
       COALESCE(deaths - lag(deaths) OVER(ORDER BY COUNTY, DATE), 0) as DailDeaths
from illinois
order by DailDeaths desc
````

### Most deaths reported by each county in IL

````python
%%sql
-- Creating the table holding valus for Illinois
WITH 
illinois_2020 AS (select * from us_counties_2020 where state = 'Illinois'),
illinois_2021 AS (select * from us_counties_2021 where state = 'Illinois'),
illinois_2022 AS (select * from us_counties_2022 where state = 'Illinois'),
illinois_2023 AS (select * from us_counties_2023 where state = 'Illinois'),
illinois AS (select * from illinois_2020 UNION select * from illinois_2021 UNION select * from illinois_2022 UNION select * from illinois_2023),
data as ( select *, ROW_NUMBER() OVER(PARTITION BY COUNTY ORDER BY DAILYDEATHS DESC) as ROWS
from 
( select *,  
       COALESCE(cases - lag(cases) OVER(ORDER BY COUNTY, DATE), 0) as DailyCases,
       COALESCE(deaths - lag(deaths) OVER(ORDER BY COUNTY, DATE), 0) as DailyDeaths
from illinois
order by DailyDeaths desc))

select * from data
where ROWS = 1
order by DailyDeaths desc
````

### Most deaths reported by each county in Wisconsin

````python
%%sql
-- Creating the table holding valus for Milwaukee in Wisconsin
WITH 
wisconsin_2020 AS (select * from us_counties_2020 where state = 'Wisconsin'),
wisconsin_2021 AS (select * from us_counties_2021 where state = 'Wisconsin'),
wisconsin_2022 AS (select * from us_counties_2022 where state = 'Wisconsin'),
wisconsin_2023 AS (select * from us_counties_2023 where state = 'Wisconsin'),
wisconsin AS (select * from wisconsin_2020 UNION select * from wisconsin_2021 UNION select * from wisconsin_2022 UNION select * from wisconsin_2023),
data as ( select *, ROW_NUMBER() OVER(PARTITION BY COUNTY ORDER BY DAILYDEATHS DESC) as ROWS
from 
( select *,  
       COALESCE(cases - lag(cases) OVER(ORDER BY COUNTY, DATE), 0) as DailyCases,
       COALESCE(deaths - lag(deaths) OVER(ORDER BY COUNTY, DATE), 0) as DailyDeaths
from wisconsin
order by DailyDeaths desc))

select * exclude(ROWS) from data
where ROWS = 1
order by DailyDeaths desc
````

### Adding Column - Region based on the state

````python
%%sql

select *, case when state in ('West Virginia', 'District of Columbia', 'Maryland', 'Virginia', 'Kentucky', 'Tennessee', 'North Carolina', 'Mississippi', 'Arkansas', 'Louisiana', 'Alabama', 'Georgia', 'South Carolina', 'Florida', 'Delaware') THEN 'SOUTH'
               when state in ('Arizona', 'New Mexico', 'Oklahoma', 'Texas') THEN 'SOUTHWEST'
               when state in ('Washington', 'Oregon', 'California', 'Nevada', 'Idaho', 'Montana', 'Wyoming', 'Utah', 'Colorado', 'Alaska', 'Hawaii') THEN 'WEST'
               when state in ('North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri', 'Wisconsin', 'Illinois', 'Michigan', 'Indiana', 'Ohio') THEN 'MIDWEST'
               when state in ('Maine', 'Vermont', 'New York', 'New Hampshire', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New Jersey', 'Pennsylvania') THEN 'NORTHEAST'
               ELSE 'NOIDEA' END as REGION
from us_counties_2020
````

### Total Deaths and Cases in Each State

````python
%%sql

WITH 
us_2020 AS (select * from us_counties_2020),
us_2021 AS (select * from us_counties_2021),
us_2022 AS (select * from us_counties_2022),
us_2023 AS (select * from us_counties_2023),
us AS (select * from us_2020 UNION select * from us_2021 UNION select * from us_2022 UNION select * from us_2023),
daily as (select *, COALESCE(cases - lag(cases) OVER(PARTITION BY STATE, COUNTY ORDER BY DATE), 0) as DailyCases, COALESCE(deaths - lag(deaths) OVER(PARTITION BY STATE, COUNTY ORDER BY DATE), 0) as DailyDeaths from us)
select state, sum(Dailycases) as TotalCases, sum(DailyDeaths) as TotalDeaths
from daily
group by state
````
